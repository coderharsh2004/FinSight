```python
import requests
from bs4 import BeautifulSoup
import time
import random
import csv
import nltk
import spacy
from newspaper import Article
from nltk.tokenize import sent_tokenize
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('vader_lexicon')

nlp = spacy.load("en_core_web_sm")
sia = SentimentIntensityAnalyzer()

# ğŸ”¹ Keywords to identify Tesla-related sentences
keywords = ["Tesla", "TSLA", "Elon Musk", "EV", "electric car", "stock", "market", "investment", "trading"]

# ğŸ”¹ Function to check if a sentence is relevant to Tesla stock
def is_relevant(text):
    if "Tesla" in text or "TSLA" in text:
        return True
    return False

# ğŸ”¹ Function to classify sentiment using Vader
def classify_sentiment(text):
    score = sia.polarity_scores(text)['compound']

    if score > 0.1 and len(positive_sentences) < 10000:
        return "Positive"
    elif score < -0.1 and len(negative_sentences) < 10000:
        return "Negative"
    elif len(neutral_sentences) < 10000:
        return "Neutral"
    return None  # Skip if one class is full

# ğŸ”¹ Function to get Tesla stock news URLs using NewsAPI
def get_news_links(api_key, num_articles=1500):
    url = f"https://newsapi.org/v2/everything?q=Tesla stock&sortBy=publishedAt&language=en&apiKey={api_key}"

    try:
        response = requests.get(url)
        data = response.json()

        if data["status"] == "ok":
            return [article["url"] for article in data["articles"][:num_articles]]
        else:
            print("Failed to fetch news links:", data)
            return []
    except Exception as e:
        print(f"Error fetching news links: {e}")
        return []
# Add delay in scraping
time.sleep(random.uniform(2, 5))
# ğŸ”¹ Function to extract text from an article
from newspaper import Article

def scrape_text(url):
    try:
        article = Article(url)
        article.download()
        article.parse()
        return article.text if "Tesla" in article.title else None  # Ensure it's Tesla-related
    except Exception as e:
        print(f"Error fetching {url}: {e}")
        return None

# ğŸ”¹ API Key for NewsAPI (Replace with your own key)
NEWSAPI_KEY = "a2bcdb25ffb4404ea911aa363e05b48c"

# ğŸ”¹ Collecting sentences by class
positive_sentences = []
negative_sentences = []
neutral_sentences = []

# ğŸ”¹ Get Tesla stock news URLs
urls = get_news_links(NEWSAPI_KEY, num_articles=1500)  # Fetch 300 articles for more data

print(f"URLs fetched: {len(urls)}")
print("First 5 URLs:", urls[:5])

for url in urls:
    text = scrape_text(url)

    if text:
        extracted_sentences = sent_tokenize(text)
        print(f"ğŸ“„ Extracted {len(extracted_sentences)} sentences from {url}")

        for sentence in extracted_sentences:
            if is_relevant(sentence):
                sentiment = classify_sentiment(sentence)
                if sentiment:
                    if sentiment == "Positive":
                        positive_sentences.append(sentence)
                    elif sentiment == "Negative":
                        negative_sentences.append(sentence)
                    elif sentiment == "Neutral":
                        neutral_sentences.append(sentence)

    if len(positive_sentences) >= 10000 and len(negative_sentences) >= 10000 and len(neutral_sentences) >= 10000:
        break  # Stop once balanced dataset is collected

# ğŸ”¹ Save data to CSV
with open("tesla_stock_sentiment_data.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Sentence", "Sentiment"])

    for sentence in positive_sentences:
        writer.writerow([sentence, "Positive"])
    for sentence in negative_sentences:
        writer.writerow([sentence, "Negative"])
    for sentence in neutral_sentences:
        writer.writerow([sentence, "Neutral"])

print(f"âœ… Scraped & labeled Tesla stock news: {len(positive_sentences)} Positive, {len(negative_sentences)} Negative, {len(neutral_sentences)} Neutral!")
```
